```python
# Calculate the total amount of waste for each column from index 531 to 1245
total_waste = waste_data_France.iloc[:, 531:1246].sum()

# Sort the total waste amounts in descending order and select the top 10
top_10_total_waste = total_waste.sort_values(ascending=False).head(10)

# Print the results
print("Top 10 columns with the highest total waste:")
print(top_10_total_waste)
```

```python
# Set the figure size for better readability
plt.figure(figsize=(12, 10))

# Create a horizontal bar chart
plt.barh(top_10_total_waste.index[::-1], top_10_total_waste.values[::-1], color='skyblue')

# Add title and labels for clarity
plt.title('Top 10 Waste Categories by Total Amount')
plt.ylabel('Waste Category')
plt.xlabel('Total Amount')

# Loop over the bars and add text annotations with the total values
for index, value in enumerate(top_10_total_waste.values[::-1]):
    plt.text(value, index, f' {value}', va='center', color='black')

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Sort the data in descending order for clarity.
sorted_sums = sums_waste_by_secteur.sort_values(ascending=True)

# Now, create a horizontal bar chart with the sorted data.
plt.figure(figsize=(10, 12))
sorted_sums.plot(kind='barh', color='skyblue')
plt.title('Total Waste Amount by Sector')
plt.xlabel('Total Waste Amount')
plt.ylabel('Sector')

# Loop over the bars and add text annotations with the total values
for index, value in enumerate(sums_waste_by_secteur.values[::-1]):
    plt.text(value, index, f' {value}', va='center', color='black')

# Use tight_layout() to adjust the layout so everything fits without overlapping
plt.tight_layout()

# Finally, show the plot.
plt.show()
```

```python
# Sort the data in descending order to have the largest sectors appear first
sorted_sums = sums_waste_by_secteur.sort_values(ascending=False)

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(sorted_sums, labels=sorted_sums.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Total Waste Amount by Sector')

# Show the plot
plt.show()
```
