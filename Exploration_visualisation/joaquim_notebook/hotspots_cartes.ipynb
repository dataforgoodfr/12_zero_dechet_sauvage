### Carte DENSITE

```python
def couleur_milieu(type):
    if type == 'Littoral (terrestre)':
        return 'lightblue'
    if type == 'Mer - Océan':
        return 'darkblue'
    if type == 'Mer - Océan':
        return 'darkblue'
    if type == 'Cours d\'eau':
        return 'cyan'
    if type == 'Zone naturelle ou rurale (hors littoral et montagne)':
        return 'green'
    if type == 'Zone urbaine':
        return 'orange'
    if type == 'Lagune et étang côtier':
        return 'red'
    if type == 'Multi-lieux':
        return 'pink'
    if type == 'Montagne':
        return 'grey'
    if type == 'Présent au sol (abandonné)':
        return 'black'
    else:
        return 'white'
```

```python
gdf = gpd.read_file(FRANCE_REGIONS)


m = folium.Map(location=[data['LIEU_COORD_GPS_Y'].mean(), data['LIEU_COORD_GPS_X'].mean()])

for index, row in data.iterrows():

    color = couleur_milieu(row['TYPE_MILIEU'])

    folium.CircleMarker(
        location=[row['LIEU_COORD_GPS_Y'], row['LIEU_COORD_GPS_X']],
        radius=row['DENSITE']*10,  
        popup=f"Densité: {row['DENSITE']} L/m2",
        color=color,
        fill=True,
        
    ).add_to(m)

m
```
